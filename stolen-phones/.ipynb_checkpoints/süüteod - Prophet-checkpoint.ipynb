```python
import pandas as pd
from fbprophet import Prophet
```

```python
df_1 = pd.read_csv('avalik_1.csv', sep='\t')
df_2 = pd.read_csv('avalik_2.csv', sep='\t')
df_3 = pd.read_csv('avalik_3.csv', sep='\t')

df = df_1.append(df_2).append(df_3)
df = df.drop_duplicates()
```

```python
df['ToimKpv'] = pd.to_datetime(df['ToimKpv'])
```

```python
df = df.groupby('ToimKpv')[['JuhtumId']].nunique()
df = df.reset_index()
df = df.rename(columns={'ToimKpv':'ds', 'JuhtumId':'y'})
```

```python
df.head()
```

```python
m = Prophet()
m.fit(df)

future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
```
