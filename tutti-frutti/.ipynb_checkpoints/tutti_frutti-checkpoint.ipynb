```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tutti_frutti_data.csv', index_col=0)
```

```python
df.info()
```

```python
df.head()
```

```python
sample_totals = df.sum(axis=1)

df_proportions = df.div(sample_totals, axis=0)
df_proportions.head()
```

```python
df_proportions.describe()
```

```python
nrows = 3; ncols = 3
num_plots = nrows*ncols

fig = plt.figure(figsize=(9, 9))
axes = [plt.subplot(nrows,ncols,i) for i in range(1,num_plots+1)]

plt.subplots_adjust(wspace=.5, hspace=.5)

for i, col in enumerate(df_proportions.columns):
    ax = axes[i]
    ax.hist(df_proportions[col])
    ax.axvline(df_proportions[col].mean(), color='#404040', linestyle='dashed', linewidth=1)
    
    ax.set_title(col)
    ax.set_ylim(0, 14)
    ax.set_xlim(0, .4)
    
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_yaxis().set_visible(False)
    
```

```python
import scipy.stats
```

```python
pop_mean = 1/9
alpha = 0.05

for col in df_proportions.columns[:9]:
    result = scipy.stats.ttest_1samp(df_proportions[col], pop_mean)
    if result[1] < alpha:
        print(f'Reject the null for {col}: t-statistic {result[0].round(4)}, p-value: {result[1].round(4)}')
    else:
        print(f'Fail to reject the null for {col}: t-statistic {result[0].round(4)}, p-value: {result[1].round(4)}')
```

```python
df_agg = pd.DataFrame()
```

```python
df_agg['greens'] = df['green_butterflies'] + df['green_frogs'] + df['green_snails']
df_agg['yellows'] = df['yellow_butterflies'] + df['yellow_frogs'] + df['yellow_snails']
df_agg['reds'] = df['red_butterflies'] + df['red_frogs'] + df['red_snails']

df_agg['butterflies'] = df['green_butterflies'] + df['yellow_butterflies'] + df['red_butterflies']
df_agg['frogs'] = df['red_frogs'] + df['yellow_frogs'] + df['red_frogs']
df_agg['snails'] = df['green_snails'] + df['yellow_snails'] + df['red_snails']
```

```python
df_agg.head()
```

```python
df_agg_proportions = df_agg.div(sample_totals, axis=0)
df_agg_proportions.head()
```

```python
df_agg_proportions.describe()
```

```python
nrows = 2; ncols = 3
num_plots = nrows*ncols

fig = plt.figure(figsize=(10, 5))
axes = [plt.subplot(nrows,ncols,i) for i in range(1,num_plots+1)]

plt.subplots_adjust(wspace=.5, hspace=.5)

for i, col in enumerate(df_agg_proportions.columns):
    ax = axes[i]
    ax.hist(df_agg_proportions[col])
    ax.axvline(df_agg_proportions[col].mean(), color='#404040', linestyle='dashed', linewidth=1)
    
    ax.set_title(col)
    ax.set_ylim(0, 7)
    ax.set_xlim(0, 1)
    
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_yaxis().set_visible(False)
```

```python
pop_mean = 1/3
alpha = 0.05
    
for col in df_agg_proportions.columns:
    result = scipy.stats.ttest_1samp(df_agg_proportions[col], pop_mean)
    if result[1] < alpha:
        print(f'Reject the null for {col}: t-statistic {result[0].round(4)}, p-value: {result[1].round(4)}')
    else:
        print(f'Fail to reject the null for {col}: t-statistic {result[0].round(4)}, p-value: {result[1].round(4)}')
```
