```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import calendar
```

```python
# 2004 - 2021
df_2004 = pd.read_excel('Tallinn-Harku_2004-2021.xlsx', skiprows=[0])
df_2004 = df_2004[['Aasta', 'Kuu', 'Päev', 'Kell (UTC)', '10 minuti keskmine tuule kiirus m/s']]
df_2004 = df_2004.rename(columns={'Aasta':'year', 'Kuu':'month', 'Päev':'day', 
                                  'Kell (UTC)':'hour', '10 minuti keskmine tuule kiirus m/s':'wind'})
df_2004['hour'] = df_2004['hour'].astype(str).str[:2].astype(int)
df_2004.head()
```

```python
# 2022
df_2022 = pd.read_csv('2022.csv')
df_2022['date'] = pd.to_datetime(df_2022['date'], format='%d.%m.%Y')
df_2022['year'] = df_2022['date'].dt.year
df_2022['month'] = df_2022['date'].dt.month
df_2022['day'] = df_2022['date'].dt.day
df_2022 = df_2022[['year', 'month', 'day', 'hour', 'Tuul kiirus (m/s)']]
df_2022 = df_2022.rename(columns={'Tuul kiirus (m/s)':'wind'})
df_2022.head()
```

```python
df = pd.concat([df_2004, df_2022], ignore_index=True)
df.head()
```

```python
df.to_csv('wind.csv', index=False)
```
