```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import googlemaps
import gmaps
```

# Data collection and cleaning

I first start by scraping each library's address from their main website.

```python
headers = requests.utils.default_headers()
headers.update({'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'})
url = 'https://keskraamatukogu.ee/kontakt/'
```

The addresses of the libraries are all stored in three separate tables so let's find them all.

```python
req = requests.get(url, headers)
soup = BeautifulSoup(req.content, 'html.parser')
html_tables = soup.select('.table-responsive > table')
```

Then let's convert each HTML table into a Pandas data frame, append each one into a list and then concatenate them all into one data frame.

```python
dfs = []

for table in html_tables:
    dfs.append(pd.read_html(str(table))[0])
    
df = pd.concat(dfs)
```

```python
df.head(3)
```

The name of a library and the address is what we need so let's drop the last two columns.

```python
df.drop(columns=[2, 3], inplace=True)
df.columns = ['name', 'address']
```

There is also other contact info in the table which are not really libraries. Let's then only keep the necessary rows - they all have 'raamatukogu' or 'kirjanduse osakond' in their name.

```python
df = df.loc[df['name'].str.contains('raamatukogu|kirjanduse osakond', regex=True)]
df.reset_index(inplace=True, drop=True)
```

```python
df.head(3)
```

Now let's proceed to get the coordinates of each library. I'll be using the Google Maps API which gives the coordinates based on the textual address.

```python
api_key = '###'
googlemaps = googlemaps.Client(key=api_key)
```

```python
df['lat'] = ''
df['long'] = ''
```

```python
for i in range(len(df['address'])):
    geocode_result = googlemaps.geocode(df['address'][i])
    df['lat'][i] = geocode_result[0]['geometry']['location']['lat']
    df['long'][i] = geocode_result[0]['geometry']['location']['lng']
```

```python
print(len(df))
df.head()
```

# Analysis

Now that we have all the necessary data we can proceed with the analytics.

First, I'll create a list of the libraries' coordinates. Each item in the list will be a tuple containing latitude and longitude. I can later use the list as an input for calculating the most optimal route.

```python
waypoints = [(df.loc[i,'lat'], df.loc[i,'long']) for i in range(len(df))]
len(waypoints)
```

I'll also create a variable which will be both the starting and ending point of the journey.

```python
home = (59.421425, 24.748945)
```

Then, I'll just let Google Maps calculate the most optimal (driving) route and plot it on a map.

```python
gmaps.configure(api_key=api_key)

fig = gmaps.figure()

origin = home
destination = home
layer = gmaps.directions.Directions(origin,
                                    destination,
                                    waypoints=waypoints,
                                    mode='driving',
                                    optimize_waypoints=True)

fig.add_layer(layer)
fig
```

Now let's also extract the length and duration of the journey.

This time, the waypoints information needs to be in a different format, so let's start with that.

```python
waypoints_dir = [str(x[0]) + ', ' + str(x[1]) for x in waypoints]   
waypoints_dir = '|'.join(waypoints_dir)
waypoints_dir = list(waypoints_dir.split(';'))
```

```python
waypoints_dir[0][:100]
```

Then, I'll write a function which extracts the data we want (duration and distance etc) as I need to use it both for walking and driving modes.

```python
def get_directions_data(mode):
    return googlemaps.directions(origin=home,
                                 destination=home,
                                 waypoints=waypoints_dir,
                                 mode=mode,
                                 optimize_waypoints=True)
```

Then, I'll create a list for each type of data we need so that I could later user those lists to put together a data frame. Then, when I query the data from Google Maps API (which returns a JSON), I'll append the necessary data into the corresponding list.

```python
dfs = []
distance_m = []
duration_s = []
start_address = []
end_address = []
start_coord = []
end_coord = []
travel_mode = []

for mode in ['driving', 'walking']:
    directions = get_directions_data(mode)

    for i, leg in enumerate(directions[0]['legs']):
        distance_m.append(leg['distance']['value'])
        duration_s.append(leg['duration']['value'])
        start_address.append(leg['start_address'])
        end_address.append(leg['end_address'])
        start_coord.append((leg['start_location']['lat'], leg['start_location']['lng']))
        end_coord.append((leg['end_location']['lat'], leg['end_location']['lng']))
        travel_mode.append(mode)
```

Now let's use the lists to create a data frame.

```python
df_itinerary = pd.DataFrame({'distance_m':distance_m,
                             'duration_s':duration_s,
                             'start_address':start_address,
                             'end_address':end_address,
                             'start_coord':start_coord,
                             'end_coord':end_coord,
                             'mode':travel_mode})
```

```python
df_itinerary.head()
```

Now let's see what are the distance and duration for both driving and walking modes.

```python
dist_drive = round(df_itinerary.loc[df_itinerary['mode']=='driving', 'distance_m'].sum()/1000, 1)
time_drive = df_itinerary.loc[df_itinerary['mode']=='driving', 'duration_s'].sum()/60

dist_walk = round(df_itinerary.loc[df_itinerary['mode']=='walking', 'distance_m'].sum()/1000, 1)
time_walk = df_itinerary.loc[df_itinerary['mode']=='walking', 'duration_s'].sum()/60

print(f'Total distance driving is {dist_drive} km and it will take {int(time_drive // 60)} h and {int(time_drive % 60)} min')
print(f'Total distance walking is {dist_walk} km and it will take {int(time_walk // 60)} h and {int(time_walk % 60)} min')

# Total distance driving is 65.9 km and it will take 2 h and 21 min
# Total distance walking is 56.6 km and it will take 11 h and 43 min
```

There is quite a difference in the distance. Let's check if the order of the waypoints in both itineraries (driving vs walking) are the same.

```python
walking_itinerary = list(df_itinerary.loc[df_itinerary['mode']=='driving', 'start_address'])
driving_itinerary = list(df_itinerary.loc[df_itinerary['mode']=='walking', 'start_address'])
walking_itinerary == driving_itinerary
```

# Visulization

Now let's also visualise both the driving and walking itineraries. I'll be using OpenStreetMap.

```python
import networkx as nx
import osmnx as ox
```

First, let's create lists for both driving and walking coordinates. Again, every item in the list is a tuple containing latitude and longitude.

```python
driving_coordinate_list = list(df_itinerary.loc[df_itinerary['mode']=='driving', 'start_coord'])
walking_coordinate_list = list(df_itinerary.loc[df_itinerary['mode']=='walking', 'start_coord'])
```

Then, let's download and model Tallinn's drivable street network.

```python
G = ox.graph_from_place('Tallinn, Estonia', network_type='drive')
```

After that, we need to convert the geo-coordinates into routes that OpenStreetMap can plot on a graph. The general idea is that the whole itinerary has to be a collection of nodes and edges that could be graphically represented on the map.

```python
def generate_routes(coordinate_list):
    # convert coordinates into nodes
    nodes_list = [ox.get_nearest_node(G, x) for x in coordinate_list]   
    nodes_list.append(nodes_list[0])

    # convert nodes into routes
    routes = []
    for i, node in enumerate(nodes_list):
        if i < len(nodes_list)-1:
            routes.append(nx.shortest_path(G, node, nodes_list[i+1], weight='length'))
    
    return routes
```

Then let's use the above function to calculate generate the routes for both driving and walking itineraries.

```python
driving_routes = generate_routes(driving_coordinate_list)
walking_routes = generate_routes(walking_coordinate_list)
```

Since I want to use different colours for visualizing driving and walking itineraries, I need to calculate how many edges both routes have.

```python
driving_edges_count = sum([len(x)-1 for x in driving_routes])
walking_edges_count = sum([len(x)-1 for x in walking_routes])
```

And now, finally, we can visualize both itineraries.

```python
ox.config(log_console=True, use_cache=True)

driving_colors = ['#e63946'] * driving_edges_count
walking_colors = ['#457b9d'] * walking_edges_count
colors = driving_colors + walking_colors

fig, ax = ox.plot_graph_routes(G, driving_routes+walking_routes, figsize=(12,12),
                               route_color=colors, route_linewidth=3, route_alpha=.5,
                               node_size=0, edge_linewidth=.5,
                               orig_dest_size=40, orig_dest_node_alpha=.25,
                               save=True, filepath='map.png')
```
